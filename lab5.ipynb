```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Question 1: Generate dataset and compute PMF & CDF
defective_counts = np.random.randint(0, 9, size=1000)  # N=1000 batches

# Compute PMF
unique, counts = np.unique(defective_counts, return_counts=True)
pmf = counts / len(defective_counts)

# Compute CDF
cdf = np.cumsum(pmf)

# Plot PMF
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.bar(unique, pmf, alpha=0.7, color='blue')
plt.xlabel("Defective Items per Batch")
plt.ylabel("Probability")
plt.title("PMF of Defective Items per Batch")

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(unique, cdf, marker='o', linestyle='-', color='red')
plt.xlabel("Defective Items per Batch")
plt.ylabel("Cumulative Probability")
plt.title("CDF of Defective Items per Batch")

plt.tight_layout()
plt.show()

# Question 2: Binomial Probability Calculation
N = 100  # Number of bulbs inspected
p = 0.03  # Probability of defect per bulb

# i) Probability of exactly 5 defective bulbs
prob_5 = binom.pmf(5, N, p)
print(f"Probability of exactly 5 defective bulbs: {prob_5:.4f}")

# ii) Probability of at least 8 defective bulbs
prob_at_least_8 = 1 - binom.cdf(7, N, p)
print(f"Probability of at least 8 defective bulbs: {prob_at_least_8:.4f}")

# iii) Expected number of defective bulbs
expected_defects = N * p
print(f"Expected number of defective bulbs: {expected_defects}")
```
